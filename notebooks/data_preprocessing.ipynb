# Preprocessing

```python
import pandas as pd
from multipride.preprocessing import clean_text
```

```python
# Import data from CSV
train_data = pd.read_csv("../data/raw/train_it.csv")
train_data
```

```python
train_data["text"] = train_data["text"].apply(lambda x: clean_text(x))

train_data["bio"] = train_data["bio"].fillna("")

train_data["bio"] = train_data["bio"].astype(str)

train_data
```

```python
train_data.to_csv("../data/processed/clean_train_data.csv", index=False)
```

We apply the same preprocessing also to the test set.

```python
# Import test data from CSV
test_data = pd.read_csv("../data/raw/test_it.csv")
test_data
```

```python
test_data["text"] = test_data["text"].apply(lambda x: clean_text(x))

test_data["bio"] = test_data["bio"].fillna("")

test_data["bio"] = test_data["bio"].astype(str)

test_data
```

```python
test_data.to_csv("../data/processed/clean_test_data.csv")
```
